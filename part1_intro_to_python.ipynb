### 6. Image manipulation with **Pillow** (PIL)
Pillow is a powerful library for image processing in Python. It allows for opening, manipulating, and saving many different image file formats.

This cell showcases the Pillow library for image processing. It opens an existing image (`example_image.jpg`), reads its dimensions, then demonstrates how to resize the image while maintaining its aspect ratio and how to convert it to grayscale. Finally, it displays the original and processed images using `matplotlib.pyplot` and cleans up the generated files and directory.

```python
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
from pathlib import Path

# Define a directory for our image operations
image_dir = Path("image_examples")
image_dir.mkdir(exist_ok=True)

# --- Use the provided image file for demonstration ---
original_image_path = Path("/content/example_image.png")

print(f"Read from:  {original_image_path}")
```

```python
# 2. Read an image
img_read = Image.open(original_image_path)
print(f"Read image with size: {img_read.size}")

# Display the original image
plt.figure(figsize=(20, 10)) # Adjusted figure size for potentially larger image
plt.subplot(1, 3, 1)
plt.imshow(img_read)
plt.title("Original Image")
plt.axis('off')
```

```python
# 3. Modify the image: Resize
# Resize to a common smaller dimension for display if needed, maintaining aspect ratio
width, height = img_read.size
max_dim = 200 # Max dimension for display
if width > height:
    new_width = max_dim
    new_height = int(height * (max_dim / width))
else:
    new_height = max_dim
    new_width = int(width * (max_dim / height))
resized_image = img_read.resize((new_width, new_height))
resized_image_path = image_dir / "resized_image.png"
resized_image.save(resized_image_path)
print(f"Resized image saved to: {resized_image_path}")

# Display the resized image
#plt.figure(figsize=(20, 10)) # Adjusted figure size for potentially larger image
plt.subplot(1, 3, 2)
plt.imshow(resized_image)
plt.title("Resized Image")
plt.axis('off')
```

```python
# 4. Modify the image: Convert to grayscale
grayscale_image = img_read.convert('LA') # 'LA' for grayscale
grayscale_image_path = image_dir / "grayscale_image.png"
grayscale_image.save(grayscale_image_path)
print(f"Grayscale image saved to: {grayscale_image_path}")

# Display the grayscale image
plt.figure(figsize=(20, 10)) # Adjusted figure size for potentially larger image
plt.subplot(1, 3, 3)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Clean up the created images and directory
print("\nCleaning up generated image files...")
if resized_image_path.exists():
    resized_image_path.unlink()
if grayscale_image_path.exists():
    grayscale_image_path.unlink()

# Remove the directory only if it's empty
try:
    image_dir.rmdir()
    print(f"Cleaned up image directory: {image_dir}")
except OSError:
    print(f"Image directory {image_dir} is not empty or does not exist, skipping removal.")
```

### 7. Video manipulation with **OpenCV** (cv2)
OpenCV is an open-source computer vision and machine learning software library. It is widely used for image and video processing tasks.

This cell demonstrates basic video processing using OpenCV. It reads an existing video file (`example_video.mp4`) frame by frame. For each frame, it converts the frame to grayscale, draws a white rectangle on it, and then writes the processed frame to a new grayscale video file (`grayscale_video.mp4`).

Taking videos from [IPN Hands](https://gibranbenitez.github.io/IPN_Hand/)

```python
import cv2
import numpy as np
from pathlib import Path
import os

# Define a directory for our video operations
video_dir = Path("video_examples")
video_dir.mkdir(exist_ok=True)

# --- Use the provided video file for demonstration --- #
input_video_path = Path("/content/videos/4CM11_19_R_#20.avi")

print(f"Read from:  {input_video_path}")
```

```python


# --- 2. Read the video, convert to grayscale, and perform an operation --- #
grayscale_output_path = video_dir / "grayscale_video.mp4"

cap = cv2.VideoCapture(str(input_video_path))

if not cap.isOpened():
    print(f"Error: Could not open video {input_video_path}")
else:
    # Get video properties for output writer
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object for grayscale output
    # Using 'mp4v' or 'XVID' codec for .mp4 files
    fourcc_gray = cv2.VideoWriter_fourcc(*'mp4v')
    out_gray = cv2.VideoWriter(str(grayscale_output_path), fourcc_gray, fps, (width, height), isColor=False) # isColor=False for grayscale

    print(f"Processing video {input_video_path}...")
    while True:
        ret, frame = cap.read() # Read a frame

        if not ret:
            break # Break the loop if no more frames

        # Convert frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Perform another basic OpenCV operation: Draw a rectangle
        cv2.rectangle(gray_frame, (50, 100), (250, 150), (255, 255, 255), 2) # White rectangle, 2px thickness

        # Write the processed grayscale frame to the output video
        out_gray.write(gray_frame)

    cap.release()
    out_gray.release()
    print(f"Processed grayscale video saved to: {grayscale_output_path}")

```

### 10. Solution

This section provides the complete solution to the image and video data pipeline analysis exercise. It integrates `pathlib` for file management, `Pillow` for image processing, `OpenCV` for video analysis, `NumPy` for numerical operations, `Pandas` for data aggregation, and `Matplotlib` for visualization.

```python
#@markdown Here you find the solution for the exercise.


import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image

# --- Setup: Define paths and create exercise directory ---
exercise_dir = Path("exercise_output")
exercise_dir.mkdir(exist_ok=True, parents=True)

original_image_path = Path("/content/example_image.png")
original_video_path = Path("/content/videos/4CM11_19_R_#20.avi")

print(f"Exercise data will be stored in: {exercise_dir}")

# --- Part 1: Image Analysis with Pillow ---
print("\n--- Part 1: Image Analysis ---")

# Load the image
img = Image.open(original_image_path)

print(f"Loaded image: {original_image_path} with size {img.size}")

# Convert the image to grayscale
grayscale_img = img.convert('L') # 'L' for grayscale

# Save the grayscale image to the exercise directory
output_grayscale_image_path = exercise_dir / "processed_grayscale_image.png"
grayscale_img.save(output_grayscale_image_path)

print(f"Saved grayscale image to: {output_grayscale_image_path}")

# Calculate the average pixel intensity of the grayscale image
average_intensity_image = np.array(grayscale_img).mean()

print(f"Average pixel intensity of image: {average_intensity_image:.2f}")


# --- Part 2: Video Analysis with OpenCV ---
print("\n--- Part 2: Video Analysis ---")

# Open the video file
cap = cv2.VideoCapture(str(original_video_path))

if not cap.isOpened():
    print(f"Error: Could not open video {original_video_path}")
else:
    frame_count = 0
    intensities = []

    print(f"Processing video {original_video_path}...")
    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Calculate the average pixel intensity of the grayscale frame
        average_intensity_frame = np.mean(gray_frame)
        intensities.append(average_intensity_frame)

        frame_count += 1
        if frame_count % 50 == 0:
            print(f"  Processed {frame_count} frames.")

    # Release the video capture object
    cap.release()
    print(f"Finished video processing. Total frames: {frame_count}")

# --- Part 3: Data Aggregation & Analysis with Pandas & NumPy ---
print("\n--- Part 3: Data Aggregation & Analysis ---")

# Create a Pandas DataFrame from the video intensity data
video_data = {'Frame': range(len(intensities)), 'Average_Intensity': intensities}
df_video_analysis = pd.DataFrame(video_data)

#print the first 5 rows of video analysis DataFrame
print("First 5 rows of video analysis DataFrame:")
display(df_video_analysis.head())

# Find the frame with the maximum average intensity using NumPy/Pandas
max_intensity_frame_idx = df_video_analysis['Average_Intensity'].idxmax()
max_intensity_value = df_video_analysis['Average_Intensity'].max()
print(f"Frame with maximum average intensity: {max_intensity_frame_idx} (Intensity: {max_intensity_value:.2f})")

# Use a conditional statement to count frames above a certain intensity threshold
intensity_threshold = np.mean(intensities) * 1.1 # 10% above overall average
frames_above_threshold = df_video_analysis[df_video_analysis['Average_Intensity'] > intensity_threshold].shape[0]
print(f"Number of frames with intensity above {intensity_threshold:.2f}: {frames_above_threshold}")


# --- Part 4: Visualization with Matplotlib ---
print("\n--- Part 4: Visualization ---")

# Plot the average intensity over frames
plt.figure(figsize=(12, 6))
plt.plot(df_video_analysis['Frame'], df_video_analysis['Average_Intensity'], color='blue')
plt.axhline(y=intensity_threshold, color='red', linestyle='--', label=f'Intensity Threshold ({intensity_threshold:.2f})')
plt.title('Average Pixel Intensity Over Video Frames')
plt.xlabel('Frame Number')
plt.ylabel('Average Pixel Intensity')
plt.grid(True)
plt.legend()
plt.show()
```
